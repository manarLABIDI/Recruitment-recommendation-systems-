```python
#effacer les colonnes unutiles
data = data.drop('Unnamed: 0.1', axis=1)
data = data.drop('Unnamed: 0', axis=1)
data.head()
```

```python
#supprimer les doublons
data.drop_duplicates()
len(data)
```

```python
#missing values
missing_values = data.isnull().sum()
len(missing_values)
missing_values[0:6]
data = data.dropna()
len(data)
```

```python
def clean_skills(skills):
    # remove any non-alphanumeric characters and split into a list of skills
    skills = re.sub(r'[^a-zA-Z0-9,\s]+', '', skills).split(',')
    # remove any leading or trailing whitespace from each skill
    skills = [skill.strip() for skill in skills]
    # remove any empty strings from the list
    skills = list(filter(None, skills))
    # return the cleaned list of skills
    return skills
def clean_skills2(list):
    return ' '.join(list)

# apply the clean_skills function to the skills column
data['required_skills'] = data['required_skills'].apply(clean_skills)
data['required_skills']= data['required_skills'].apply(clean_skills2)


def clean_text(text):
    # remove any non-alphanumeric characters and replace with a space
    text = re.sub(r'[^a-zA-Z0-9]+', ' ', text)
    # remove any leading or trailing whitespace from the text
    text = text.strip()
    # return the cleaned text
    return text

# apply the clean_text function to the title and graduation columns
data['title'] = data['title'].apply(clean_text)
data['graduation'] = data['graduation'].apply(clean_text)

experiences=data["experience"]
# Créer un dictionnaire pour stocker les années d'expérience et les numéros correspondants
dict_exp = {}
num_exp = 0
for experence in experiences:
    if experence not in dict_exp:
        dict_exp[experence] = num_exp
        num_exp += 1

# Remplacer les valeurs textuelles par les numéros correspondants
data.replace({"experience": dict_exp}, inplace=True)

# Afficher le DataFrame modifié
data.head(100)


```

```python
## Download the NLTK modules(natural language toolkit)
import nltk 
nltk.download('punkt') #"punkt" est un module qui fournit un tokenizer de phrases (séparation du texte en phrases individuelles).
nltk.download('averaged_perceptron_tagger') #"averaged_perceptron_tagger" est un module qui fournit un tagger de parties du discours (POS) pour l'anglais.
nltk.download('wordnet') #"wordnet" est un module qui fournit un dictionnaire lexical de l'anglais qui est utilisé pour la lemmatisation.

from nltk.stem import WordNetLemmatizer #La lemmatisation est le processus de conversion d'un mot en sa forme de base, ou "lemme". 
lemmatizer = WordNetLemmatizer() 

from nltk.corpus import stopwords #liste des mots vides(a , an, at,and,in,on..)
nltk.download('stopwords') 
stop_words = set(stopwords.words('english')) 
#La variable VERB_CODES est un ensemble de chaînes représentant les codes de parties du discours des verbes en anglais. Ces codes de parties du discours de Penn Treebank (PTB) sont utilisés pour identifier les verbes dans le texte, afin de pouvoir les lemmatiser correctement. Les codes de parties du discours inclus dans VERB_CODES sont: VB (verbe de base), VBD (passé simple), VBG (participe présent), VBN (participe passé), VBP (présent non-3ème personne) et VBZ (présent 3ème personne).
VERB_CODES = {'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'}


## Creating the text manipulation function
def preprocess_sentences(text): 

    text = text.lower() 
    temp_sent =[] 
    words = nltk.word_tokenize(text) 
    tags = nltk.pos_tag(words) 
    
    for i, word in enumerate(words): 
        if tags[i][1] in VERB_CODES:   
            lemmatized = lemmatizer.lemmatize(word, 'v') #'v' lemmatizer les verbes
        else: 
            lemmatized = lemmatizer.lemmatize(word) #par défaut 'n' lemmatizer les noms
        if lemmatized not in stop_words and lemmatized.isalpha(): 
            temp_sent.append(lemmatized) 

    finalsent = ' '.join(temp_sent) 
    finalsent = finalsent.replace("n't", " not") 
    finalsent = finalsent.replace("'m", " am") 
    finalsent = finalsent.replace("'s", " is") 
    finalsent = finalsent.replace("'re", " are") 
    finalsent = finalsent.replace("'ll", " will") 
    finalsent = finalsent.replace("'ve", " have") 
    finalsent = finalsent.replace("'d", " would") 

    return finalsent 


## Creating a new column that we applied the text manipulation function 
data["title"] = data["title"].apply(preprocess_sentences) 
data["required_skills"]=data["required_skills"].apply(preprocess_sentences)
data["graduation"]=data["graduation"].apply(preprocess_sentences)
data.head(100)
```
