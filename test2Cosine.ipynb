```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

cleanedR_data = cleanedR_data.dropna()
# Creating a CountVectorizer object
cv = CountVectorizer()

# Creating the feature matrices for each category
count_title = cv.fit_transform(cleanedR_data.title)
count_skills = cv.fit_transform(cleanedR_data.required_skills)
count_graduation = cv.fit_transform(cleanedR_data.graduation)
count_experience = cv.fit_transform(cleanedR_data.experience)



```

```python

# Creating TfidfTransformer objects and transforming feature matrices
tfidf_transformer = TfidfTransformer()
title_tfidf = tfidf_transformer.fit_transform(count_title)
skills_tfidf = tfidf_transformer.fit_transform(count_skills)
graduation_tfidf = tfidf_transformer.fit_transform(count_graduation)
experience_tfidf = tfidf_transformer.fit_transform(count_experience)
```

```python
# Creating data frames for each TF-IDF transformed feature matrix
table_title_tfidf = title_tfidf.todense()
df_title_tfidf = pd.DataFrame(table_title_tfidf, 
                              columns=cv.get_feature_names(), 
                              index=cleanedR_data.index)

table_skills_tfidf = skills_tfidf.todense()
df_skills_tfidf = pd.DataFrame(table_skills_tfidf, 
                               columns=cv.get_feature_names(), 
                               index=cleanedR_data.index)

table_graduation_tfidf = graduation_tfidf.todense()
df_graduation_tfidf = pd.DataFrame(table_graduation_tfidf, 
                                   columns=cv.get_feature_names(), 
                                   index=cleanedR_data.index)

table_experience_tfidf = experience_tfidf.todense()
df_experience_tfidf = pd.DataFrame(table_experience_tfidf, 
                                    columns=cv.get_feature_names(), 
                                    index=cleanedR_data.index)
print("Experience TF-IDF Feature Vector:\n", df_experience_tfidf)


```
